```python
from matplotlib import pyplot as plt
from fracturetoughness import FractureToughness
from threepointbend import ThreePointBend
```

```python
# cherry-pick examples for display plots

# notched samples
notched_cast_4 = FractureToughness('Thesis_Samples_Data/Notched_cast_4/Notched_cast_4.csv', 90, 25.87, 32.31, 0.97, cmp=1.99)
notched_solid_4 = FractureToughness('Thesis_Samples_Data/Notched_Solid_4/Notched_Solid_4.csv', 90, 25.08, 41.76, 0.97, cmp=1.86)
notched_layered_4 = FractureToughness('Thesis_Samples_Data/Notched_Layered_4/Notched_Layered_4.csv', 90, 23.1, 43.76, 0.97, cmp=1.48, multi=True)

notched_cast_4.display_peaks()
notched_solid_4.display_peaks()
notched_layered_4.display_peaks()
```

```python
notched_cast_4.do_analysis(2379)
notched_solid_4.do_analysis(2063)
notched_layered_4.do_analysis(2245, stop_index=18755) #18755
```

```python
# unnotched samples
flexural_cast_3 = ThreePointBend('Thesis_Samples_Data/Flexural_Cast_3/Flexural_Cast_3.csv', 90, 28.42, 32.18, cmp=2.01)
flexural_solid_2 = ThreePointBend('Thesis_Samples_Data/Flexural_Solid_2/Flexural_Solid_2.csv', 90, 26.37, 39.97, cmp=1.83)
flexural_layered_2 = ThreePointBend('Thesis_Samples_Data/Flexural_Layered_2/Flexural_Layered_2.csv', 90, 23.04, 43.37, cmp=1.49, multi=True)
flexural_rebar_2 = ThreePointBend('Thesis_Samples_Data/Flexural_Rebar_2/Flexural_Rebar_2.csv', 90, 26.01, 47.68, cmp=1.75, multi=True)
```

```python
flexural_cast_3.display_peaks()
flexural_solid_2.display_peaks()
flexural_layered_2.display_peaks()
flexural_rebar_2.display_peaks()
```

```python
flexural_cast_3.do_analysis(2198)
flexural_solid_2.do_analysis(2034)
flexural_layered_2.do_analysis(4250, stop_index=15000)
flexural_rebar_2.do_analysis(4107, stop_index=81000)
```

```python
plt.figure(figsize=(16, 8))

plt.suptitle('Corresponding Cracks in Samples across Load-displacement and Load-CMOD', fontsize=24)

sub = plt.subplot(1, 2, 1)
notched_cast_4.plot('crosshead_linearized', 'load_linearized_normalized', sub, label='Cast Cement', c='b', mc='#3399ff')
notched_solid_4.plot('crosshead_linearized', 'load_linearized_normalized', sub, label='3DP Cement', c='#ff9933', mc='#ff2f2f')
notched_layered_4.plot('crosshead_linearized', 'load_linearized_normalized', sub, label='3DP Cement-Silicone', c='#99ff33', mc='#36a841')
plt.xlabel('Crosshead Displacement (mm)', fontsize=16)
plt.ylabel('Specific Load', fontsize=16)
plt.legend(fontsize=12)
plt.title('Specific Load vs Crosshead', fontsize=22)


sub = plt.subplot(1, 2, 2)
notched_cast_4.plot('extenso', 'load_linearized_normalized', sub, label='Cast Cement', c='b', mc='#3399ff')
notched_solid_4.plot('extenso', 'load_linearized_normalized', sub, label='3DP Cement', c='#ff9933', mc='#ff2f2f')
notched_layered_4.plot('extenso', 'load_linearized_normalized', sub, label='3DP Cement-Silicone', c='#99ff33', mc='#36a841')
plt.xlabel('CMOD (mm)', fontsize=16)
plt.ylabel('Specific Load', fontsize=16)
plt.legend(fontsize=12)
plt.title('Specific Load vs CMOD', fontsize=22)
```

```python
plt.figure(figsize=(15, 10))

plt.suptitle('Specific Load vs CMOD', fontsize=24)

sub = plt.subplot(1, 1, 1)
notched_cast_4.plot('extenso', 'load_linearized_normalized', sub, label='Cast Cement', c='b', mc='#3399ff')
notched_solid_4.plot('extenso', 'load_linearized_normalized', sub, label='3DP Cement', c='#ff9933', mc='#ff2f2f')
notched_layered_4.plot('extenso', 'load_linearized_normalized', sub, label='3DP Cement-Silicone', c='#99ff33', mc='#36a841')
plt.xlabel('CMOD (mm)', fontsize=16)
plt.ylabel('Load', fontsize=16)
plt.legend(fontsize=12)
```

```python
plt.figure(figsize=(8, 6))

plt.suptitle('R-Curve: Kjc vs Crack Extension', fontsize=24)

sub = plt.subplot(1, 1, 1)
# notched_cast_4.plot('a_CMOD', 'K_jc', sub, label='Cast Cement', c='b', mc='#3399ff')
notched_solid_4.plot('a_CMOD', 'K_jc', sub, label='3DP Cement', c='#ff9933', mc='#ff2f2f')
notched_layered_4.plot('a_CMOD', 'K_jc', sub, label='3DP Concrete-Cement', c='#99ff33', mc='#36a841')

plt.plot(0, notched_cast_4.K_jc[0], label='Cast Cement', color='c', markerfacecolor='b', markeredgewidth=0, marker='.', markersize=10)
         
plt.axvline(x=5.25, ymin=0.01, ymax = 0.99, color='#202020', linestyle='dotted', label='ASTM E1820 Limit')

plt.xlabel('Crack Extension (mm)', fontsize=16)
plt.ylabel('Specific Kjc', fontsize=16)
plt.legend(fontsize=12, loc='best')

```
